```python
house_data = house_data.drop(['date'], axis= 1)
house_data
```

```python
X = house_data[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'condition', 'grade', 'sqft_above', 'sqft_basement',
       'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15',
       'sqft_lot15']]
y = house_data['price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y, test_size = 0.3)
```

```python
lm = LinearRegression()
```

```python
Xtrain
```

```python
ytrain
```

```python
lm.fit(Xtrain, ytrain)
```

```python
y_prediction = lm.predict(Xtest)
y_prediction
```

```python
df_results = pd.DataFrame({'Actual': ytest, 'Predicted': y_prediction})
df_results['Residuals'] = abs(df_results['Actual']) - abs(df_results['Predicted'])
df_results
```

```python
from sklearn import metrics

print('MSE', metrics.mean_squared_error(ytest, y_prediction))
print('\nRMSE', np.sqrt(metrics.mean_squared_error(ytest, y_prediction)))
print('\nAccuracy', metrics.explained_variance_score(ytest, y_prediction))
```
